```python
# 1.Find the least amount sale that was done for each item. 
a=pd.read_excel('B:/r and python ass/SaleData.xlsx') 
df2= a.loc[a.groupby('Item')['Sale_amt'].idxmin()]
print(df2.get(['Item','Sale_amt']))
a.head()
```

```python
# 2.Compute the total sales for each year and region across all items 
a['Order'] = pd.DatetimeIndex(a['OrderDate']).year
df=a.groupby(['Order','Region','Item'])
df.sum()
```

```python
# 3.Create new column 'days_diff' with number of days difference between reference date passed and each order date
a['days_diff']='2019-12-03'
a['days_diff']= pd.to_datetime(a['days_diff']) 
a['days_diff']= a['days_diff']-a['OrderDate']
a.head()
```

```python
# 4.Create a dataframe with two columns: 'manager', 'list_of_salesmen'. Column 'manager' will contain the unique managers present and column 'list_of_salesmen' will contain an array of all salesmen under each manager.

df1=a.groupby('Manager')['SalesMan'].unique().reset_index(name='list_of_salesmen')
df1
```

```python
# 5.For all regions find number of salesman and total sales. Return as a dataframe with three columns Region, salesmen_count and total_sales

df=a.groupby('Region')['SalesMan'].apply(set).reset_index(name='list_of_salesmen')
df['salesman_count']=df['list_of_salesmen'].str.len()
df3=a.groupby('Region')['Sale_amt'].apply(sum).reset_index(name='sum')
df['total_sales']=df3['sum']
df
```

```python
# 6.Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales 
df6=a.groupby('Manager')['Sale_amt'].apply(sum).reset_index(name='percent_sales')
s=df6['percent_sales'].sum()
df6['percent_sales']=(df6['percent_sales']/s)*100 
df6
```
