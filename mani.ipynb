```python
df2=pd.read_csv('C:\\Users\\mohana.kothuri\\Downloads\\imdb.csv',escapechar="\\")
```

```python
df2.head()
```

```python
df3=pd.pivot_table(df2, index= ['year'], values=['imdbRating'], aggfunc= ['mean']).round(2)
```

```python
df2=df2.dropna()
#df3=pd.pivot_table(df2, index= ['year'], values=['imdbRating'], aggfunc= list)
df3['list_of_ratings']=df2.groupby('year')['imdbRating'].apply(list)
df3.sort_values('year')
df3.head(19)
```

```python

df3['top_10_ratings']=df3['list_of_ratings'].apply(lambda i:sorted(i)[-((len(i)//10) if len(i)>10 else 1):])
df3['number of movies']=df3['top_10_ratings'].apply(len)
df3['mean imdbRating']=df3['top_10_ratings'].apply(np.mean)
df3.head(30)
```
